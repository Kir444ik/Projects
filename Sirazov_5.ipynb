```python
import math
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import tree, metrics
```

```python
#Задание:
#Создайте класс по работе с тригонометрическими функциями. В классе
#должны быть реализованы функции вычисления:
#− косинуса;
#− синуса;
#− тангенса;
#− арксинуса;
#− арккосинуса;
#− арктангенса;
#− перевода из градусов в радианы.

class Trig:
    def __init__(self, deg):
        self.deg = deg
    
    
    def degtorad(self):
        return self.deg*0.017
    
    def sin(self):
        return round(math.sin(self.degtorad()), 1)
    
    def cos(self):
        return round(math.cos(self.degtorad()), 1)
    
    def tg(self):
        return round((math.sin(self.degtorad())/math.cos(self.degtorad())), 1)
    
    def arcsin(self):
        return round(math.asin(self.degtorad()), 1)
    
    def arccos(self):
        return round(math.acos(self.degtorad()), 1)
    
    def arctg(self):
        return round(math.atan(self.degtorad()), 1)
    
d = Trig(10)
print(d.sin())
print(d.cos())
print(d.tg())
print(d.arcsin())
print(d.arccos())
print(d.arctg())
print("New trig","\n")
d = Trig(0)
print(d.sin())
print(d.cos())
print(d.tg())
print(d.arcsin())
print(d.arccos())
print(d.arctg())
```

```python
#Задача:
#Постройте классификатор на основе дерева принятия решений
#следующего датасета:


dataset = pd.DataFrame(np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]]))
target = [0, 0, 0, 1, 1, 1]
x_train, x_test, y_train, y_test = train_test_split(dataset, target, test_size=0.2)
print(x_train, y_train)
classifier = DecisionTreeClassifier()
classifier.fit(x_train, y_train)
tree.plot_tree(classifier)
y_pred = classifier.predict(x_test)
print(x_train, y_train ,"\n", y_pred,"\n",confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Задание. Постройте модель регрессии для данных из предыдущей рабочей
#тетради.Для примера можно взять потребления газа (в миллионах
#галлонов) в 48 штатах США или набор данных о качестве красного вина:
#https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/pet
#rol_consumption.csv
#https://raw.githubusercontent.com/aniruddhachoudhury/Red-WineQuality/master/winequality-red.csv
#Постройте прогноз. Оцените точность модели.
from sklearn.tree import DecisionTreeRegressor

url = r'https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv'
dataset = pd.read_csv(url)
print(dataset.head())
print(dataset.shape)
print(dataset.describe())
plt.scatter(dataset['fixed acidity'], dataset['volatile acidity'], color='b', label = 'Заработная плата')
plt.xlabel('Опыт(лет)')
plt.ylabel('Зарплата')
plt.show()
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(x)
print(y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 0)
regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)
tree.plot_tree(regressor)
y_pred = regressor.predict(x_test)
print(y_pred)
df = pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
print(df)
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
metrics.mean_absolute_error(y_test, y_pred)/np.average(y)*100
```
